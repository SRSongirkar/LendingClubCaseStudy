# Data Cleaning

```python
# Percentage wise NA by column name
lc_data.isnull().sum()*100/len(lc_data)
```

```python
# Filetring the column where NA value is less than 50%
lc_data_2 = lc_data.loc[:, lc_data.isin(['NULL',np.nan]).mean() < .5]
```

```python
# Checking variable percentage of 0 or 0.0
lc_data_2.isin([0,0.0]).mean()
```

```python
# Removing columns where percentage of 0 or 0.0 are greater than 80%
lc_data_3 = lc_data_2.loc[:, lc_data_2.isin([0, 0.0]).mean() < .8]
```

```python
# Checking head of lc_data_3
lc_data_3.head(3)
```

```python
# Checking Null values & data type
lc_data_3.info()
```

```python
# Number of unique ids
print("Number of unique ids: ", lc_data_3.id.nunique())
#39717

# Number of unique ids
print("Number of unique ids: ", lc_data_3.member_id.nunique())
#39717

# Number of records
print("Number of records: ", len(lc_data_3))
#39717

# As number of ids == number of member ids == total number of records == 39717
# So no use of keeping "id" & "member_id" column in dataset

# Removing "id" & "member_id"
lc_data_3.drop(['id','member_id'], axis = 1, inplace = True)

lc_data_3.head(3)
```

```python
# Checking type of loan status
lc_data_3.loan_status.value_counts()

#Fully Paid     32950
#Charged Off     5627
#Current         1140

# As we cant make any conclusive outcome regarding Defaulter or Non Defatulter.
# So, better to filter out rows == "Current"

lc_data_4 = lc_data_3[lc_data_3.loan_status != "Current"]
```
